```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

y=np.random.normal(loc=0.5,scale=0.4,size=1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))

plt.figure()

plt.subplot(221)
plt.plot(x,y)
plt.yscale('linear')
plt.title('liner')
plt.grid()

plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid()

plt.subplot(223)
plt.plot(x,y)
plt.yscale('symlog')
plt.title('symlog')
plt.grid()

plt.subplot(224)
plt.plot(x,y)
plt.yscale('logit')
plt.title('logit')
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from mpl_toolkits.axes_grid1 import Grid

def example_plot(ax,fontsize=12):
  ax.plot([1,2])
  ax.set_xlabel('xlabel',fontsize=fontsize)
  ax.set_ylabel('ylabel',fontsize=fontsize)
  ax.set_title('title',fontsize=fontsize)

plt.close('all')

fig=plt.figure()
grid = Grid(fig, rect=111, nrows_ncols=(2,2),
            axes_pad=0.25, label_mode='L',
            )

for ax in grid:
    example_plot(ax)
    ax.title.set_visible(False)

plt.tight_layout()
plt.show()
```
